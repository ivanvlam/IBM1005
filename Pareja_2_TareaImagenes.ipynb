### Pregunta 2

Calcule la transformada de Radon del circulo unitario defindo como

$$
f(x, y) = \begin{cases}
            0 \text{ si } x^2 + y^2 > 1 \\
            1 \text{ si } x^2 + y^2 \le 1
          \end{cases}
$$

#### Solucion

Observamos que la funcion es circularmente simetrica por lo que solo necesitamos calcular para un angulo (arbitrariamente eligiremos $\theta=0$. 

![alt text](https://drive.google.com/uc?export=view&id=1qKjVbU-XSN95FAeqxiXn5IyDvX3tCaOW)

Como indica la figura necesitamos calcular la siguiente integral

$$
g_\theta(r) = \int^a_bf(x, y)dx
$$

La funcion es 1 dentro de interbalo de integracion, luego 

$$
g_\theta(r) = \int^a_bdx
$$

Por lo tanto nos basta con conocer los valores de a y b. Dado que es un circulo unitario tenemos,

$$
g_\theta(r) = \int^{\sqrt{1 - r^2}}_{-\sqrt{1 - r^2}}dx
$$

Luego 

$$
g_\theta(r) = \begin{cases}
              2\sqrt{1-r^2} \text{ si } |r| \le 1 \\
              0 \text{ e.o.c}
          \end{cases}
$$


###  Pregunta 3

Calcule usando [sikit-image](https://scikit-image.org/docs/dev/auto_examples/transform/plot_radon_transform.html) la transformada de la pregunta anterior. Comparela con la calculada analiticamente.

#### Solucion

```python
import numpy as np
import matplotlib.pyplot as plt

# creamos una grilla con valores de x e y
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

xx, yy = np.meshgrid(x, y)
```

```python
def sphere(x, y, r):
    return 1*(x**2 + y**2 < 1)

s = sphere(xx, yy, 1)

plt.imshow(s, cmap='gray')
```

```python
from skimage.transform import radon, rescale
theta = np.linspace(0., 180., 100, endpoint=False)
sinogram = radon(s, theta=theta, preserve_range=True)

plt.imshow(sinogram, cmap='gray')
```

```python
# comparemos con la encontrada analiticamente
r = np.linspace(-2, 2, 100)
g = np.zeros_like(r)
g = np.where(np.abs(r) < 1, 2*(np.sqrt(1 - r**2)), 0 )

plt.title("Transformada de Radon de la imagen")
plt.plot(sinogram[:, 0])
plt.show()
plt.title("Transformada de Radon analitica")
plt.plot(r, g)
```

Calcule anlaliticamente la transofrmada de radon de la siguiente figura.

Ayuda: Considere la transformada de RadÃ³n del cÃ­rculo unitario de la parte formativa y utilice el hecho que la transformada de RadÃ³n es **lineal**.

$F: V \rightarrow W$ es una transformada lineal si cumple que 

1. $F(u+v)=F(u)+F(v) \quad \forall u, v \in V$
2. $F(k . v)=k . F(v) \quad \forall v \in V, \quad \forall k \in \mathbb{R}$

```python
import numpy as np
import matplotlib.pyplot as plt

# creamos una grilla con valores de x e y
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

xx, yy = np.meshgrid(x, y)
```

```python
def disco(x, y, r):
    return 1*(x**2 + y**2 < 1) & 1*(x**2 + y**2 > 1/4)

s = disco(xx, yy, 1)

plt.imshow(s, cmap='gray')
```

```python
from skimage.transform import radon, rescale
theta = np.linspace(0., 180., 100, endpoint=False)
sinogram = radon(s, theta=theta, preserve_range=True)

plt.imshow(sinogram, cmap='gray')
```

```python
r = np.linspace(-2, 2, 100)

plt.title("Transformada de Radon del disco")
plt.plot(sinogram[:, 0])
plt.show()
```

```python
def circulo_grande(x, y, r):
    return 1*(x**2 + y**2 < 1)

s = circulo_grande(xx, yy, 1)

plt.imshow(s, cmap='gray')
```

```python
from skimage.transform import radon, rescale
theta = np.linspace(0., 180., 100, endpoint=False)
sinogram_grande = radon(s, theta=theta, preserve_range=True)

plt.imshow(sinogram_grande, cmap='gray')
```

```python
def circulo_chico(x, y, r):
    return 1*(x**2 + y**2 < 1/4)

s = circulo_chico(xx, yy, 1/2)

plt.imshow(s, cmap='gray')
```

```python
from skimage.transform import radon, rescale
theta = np.linspace(0., 180., 100, endpoint=False)
sinogram_chico = radon(s, theta=theta, preserve_range=True)

plt.imshow(sinogram_chico, cmap='gray')
```

```python
r = np.linspace(-2, 2, 100)
g = np.zeros_like(r)
g1 = np.where(np.abs(r) < 1, 2*(np.sqrt(1 - r**2)), 0 )
g2 = np.where(np.abs(r) < 1/2, 2*(np.sqrt(1/4 - r**2)), 0 )

fig, ax = plt.subplots(2, 2, figsize = (12, 12))

ax[0][0].set_title("Transformada de Radon del circulo grande")
ax[0][0].plot(sinogram_grande[:, 0])

ax[0][1].set_title("Transformada de Radon del circulo chico")
ax[0][1].plot(sinogram_chico[:, 0])

ax[1][0].set_title("Transformada de Radon analitica cÃ­rculo grande")
ax[1][0].plot(r, g1)

ax[1][1].set_title("Transformada de Radon analitica cÃ­rculo chico")
ax[1][1].plot(r, g2)

fig.show()
```

```python
r = np.linspace(-2, 2, 100)
g = np.zeros_like(r)

fig, ax = plt.subplots(2, 1, figsize = (8, 12))

ax[0].set_title("Transformada de Radon resta de cÃ­rculos")
ax[0].plot(sinogram_grande[:, 0] - sinogram_chico[:, 0])

ax[1].set_title("Transformada de Radon analitica")
ax[1].plot(r, g1 - g2)

fig.show()
```

A partir de los grÃ¡ficos, se puede apreciar que mediante la representaciÃ³n grÃ¡fica del disco y la de la resta de cÃ­rculos concentricos se generÃ³ la misma transformada de RadÃ³n, lo que es de esperarse, debido a la linealidad de esta.
